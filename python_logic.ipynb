```python
class Pins:
    def __init__(self):
        # State: Down - 0; Up - 1
        # default state is gonna be all Down
        self.pins = {
            1: {"state": 0},
            2: {"state": 0},
            3: {"state": 0},
            4: {"state": 0}
        }
    
    def change_state(self, pin_number, state):
        self.pins[self.pin_number] = self.state
```

```python
class Clock:
    def __init__(self):
        self.clocks = {
            (0, 0, 0): {"type": "Corner",
                        "time": 12,
                        "quadrant": {1}
                        },
            (0, 1, 0): {"type": "Edge",
                        "time": 12,
                        "quadrant": {1, 2}
                        },
            (0, 2, 0): {"type": "Corner",
                        "time": 12,
                        "quadrant": {2}
                        },
            (1, 0, 0): {"type": "Edge",
                        "time": 12,
                        "quadrant": {1, 3}
                        },
            (1, 1, 0): {"type": "Centre",
                        "time": 12,
                        "quadrant": {1, 2, 3, 4}
                        },
            (1, 2, 0): {"type": "Edge",
                        "time": 12,
                        "quadrant": {2, 4}
                        },
            (2, 0, 0): {"type": "Corner",
                        "time": 12,
                        "quadrant": {3}
                        },
            (2, 1, 0): {"type": "Edge",
                        "time": 12,
                        "quadrant": {3, 4}
                        },
            (2, 2, 0): {"type": "Corner",
                        "time": 12,
                        "quadrant": {4}
                        },
            (0, 0, 1): {"type": "Corner",
                        "time": 12,
                        "quadrant": {2}
                        },
            (0, 1, 1): {"type": "Edge",
                        "time": 12,
                        "quadrant": {1, 2}
                        },
            (0, 2, 1): {"type": "Corner",
                        "time": 12,
                        "quadrant": {1}
                        },
            (1, 0, 1): {"type": "Edge",
                        "time": 12,
                        "quadrant": {2,4}
                        },
            (1, 1, 1): {"type": "Centre",
                        "time": 12,
                        "quadrant": {1, 2, 3, 4}
                        },
            (1, 2, 1): {"type": "Edge",
                        "time": 12,
                        "quadrant": {1, 3}
                        },
            (2, 0, 1): {"type": "Corner",
                        "time": 12,
                        "quadrant": {4}
                        },
            (2, 1, 1): {"type": "Edge",
                        "time": 12,
                        "quadrant": {3, 4}
                        },
            (2, 2, 1): {"type": "Corner",
                        "time": 12,
                        "quadrant": {3}
                        }
        }
    
    def add_time(initial_time, move):
        return ((initial_time - 1 + move) % 12) + 1
    
    def change_time(self, row, column, face, initial_time, move):
        self.clocks[(row, column, face)]['time'] = self.add_time(self.initial_time, self.move)
```

```python
pins = {
    1: {"state": 0},
    2: {"state": 1},
    3: {"state": 0},
    4: {"state": 0}
}

main_side = 0
wheel = 2
move = 3

quad_pins = set()

for i in range(1, 5):
    if pins[i]['state'] == 1:
        quad_pins.add(i)

full_set = {1, 2, 3, 4}
```

```python
for k,v in clocks.items():
    if k[2] == main_side:
        if wheel in quad_pins:
            if v['quadrant'] & (quad_pins):
                v['time'] = add_time(v['time'], move)
        else:
            if v['type'] == 'Corner' and not (v['quadrant'] & quad_pins):
                v['time'] = add_time(v['time'], move)
    
quad_pins = full_set - quad_pins
for k,v in clocks.items():
    if k[2] != main_side:
        if wheel in quad_pins:
            if v['quadrant'] & (quad_pins):
                v['time'] = add_time(v['time'], -move)
        else:
            if v['type'] == 'Corner' and not (v['quadrant'] & quad_pins):
                v['time'] = add_time(v['time'], -move)
```
